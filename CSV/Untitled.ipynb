```python
import pandas as pd
```

```python
loan = pd.read_csv('./loan.csv')
loan.head()
```

```python
loan.loc[0, '상품특징'][:-1]
```

```python
for idx, l in enumerate(loan['상품특징']):
    try:
        if l[-1] != '.':
            if l[-1] == '!':
                loan.loc[idx, '상품특징'] = loan.loc[idx, '상품특징'][:-1]
                loan.loc[idx, '상품특징'] += '입니다.'
            else:
                loan.loc[idx, '상품특징'] += '입니다.'
    except:
        continue
```

```python
for l in loan['상품특징']:
    print(l)
```

```python
loan.to_csv('./loanFinal.csv', index = False)
```
